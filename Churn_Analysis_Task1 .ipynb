```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#data cleaning
data=pd.read_csv("churn-dataset.csv")
```

```python
sns.countplot(x='Churn',data=data)
plt.title("churn vs non churn ")
plt.show()
```

```python


sns.countplot(x='State', hue='Churn', data=data)
plt.title('Churn by State')
plt.xticks(rotation=70)
plt.show()

```

```python
plt.figure(figsize=(10,8))
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

data.columns
```

```python
#customer service calls and churn 
data.groupby('Churn')['Customer service calls'].mean()
```

```python
#top 5 staes with hightest churns
top_states = data[data['Churn'] == True]['State'].value_counts().head(5)
print(top_states)
```

```python
#High Total Day Minutes is a strong churn signal.
data.groupby('Churn')['Total day minutes'].mean()
```

```python
# how many customer churn vs stayed
churn_counts = data['Churn'].value_counts()
labels = ['Not Churned', 'Churned']
colors = ['#66b3ff','#ff9999']

plt.pie(churn_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Customer Churn Rate')
plt.show()
```

```python
# churn by customer service call 
churn_by_calls = data.groupby('Customer service calls')['Churn'].mean() * 100

churn_by_calls.plot(kind='line', marker='o')
plt.xlabel('Customer Service Calls')
plt.ylabel('Churn Rate (%)')
plt.title('Churn Rate vs Customer Service Calls')
plt.grid()
plt.show()
```

```python
#total day minute vs service calls 
sns.scatterplot(x='Total day minutes', y='Customer service calls', hue='Churn', data=data)
plt.title('Total Day Minutes vs Customer Service Calls')
plt.show()
```
